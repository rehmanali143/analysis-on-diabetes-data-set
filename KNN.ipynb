```python
X=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
X      #here in x we take features 
```

```python
y=df['Outcome']        #in y we take predictions
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.40,random_state=42)
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
y_test
```

```python
scaler=StandardScaler()
```

```python
scaled_X_train=scaler.fit_transform(X_train)
```

```python
scaled_X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model =KNeighborsClassifier(n_neighbors=27)
```

```python
knn_model.fit(scaled_X_train,y_train)
```

```python
KNeighborsClassifier(n_neighbors=27)
```

```python
prediction = knn_model.predict(scaled_X_test)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score    #sklearn.metrics  is module used for accuracy 
score = accuracy_score(y_test,prediction)
score
```
